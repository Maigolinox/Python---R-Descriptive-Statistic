## TEXTO EN MATPLOTLIB

```python
mu=100
sigma=20
x=mu+sigma*np.random.randn(10000)
```

n es el numero de elementos del histogramaa,bins la cantidad de divisiones que hace d elos datos,patchhes conjuntos de datos respectivamente

```python
n,bins,patches=plt.hist(x,50,facecolor="g",alpha=0.8)
```

```python
n,bins,patches=plt.hist(x,50,facecolor="g",alpha=0.8,density=True)
```

```python
n,bins,patches=plt.hist(x,50,facecolor="g",alpha=0.8,density=True)
plt.xlabel("Coeficiente Intelectual")
plt.ylabel("Probabilidad")
plt.title("Histograma de coeficiente intelectual")
```

```python
n,bins,patches=plt.hist(x,50,facecolor="g",alpha=0.8,density=True)
plt.xlabel("Coeficiente Intelectual")
plt.ylabel("Probabilidad")
plt.title("Histograma de coeficiente intelectual",fontsize=18)
plt.text(40,0.015,r'$\mu=100,\ \sigma=20$')#Coordenada 40,0.015
#para colocar texto de LaTeX es necesario colocar una r antes de la cadena de texto
plt.axis([25,180, 0,0.025])#x de 25 a 180 y de 0 a 0.025
plt.grid(True,alpha=0.3)

plt.show()
```

```python
#como colocar una flechita con texto anotado
plt.figure(figsize=(10,6),dpi=90)#se define el tamaño del dibujo en pulgadas
plt.subplot(1,1,1)#1 fila 1 columna 1 er gráfico
x=np.arange(0,10*np.pi,0.01)
y=np.cos(x)
plt.plot(x,y,lw=2.0)
plt.annotate("Maximo local",xy=(4*np.pi,1),xytext=(15,1.5),arrowprops=dict(facecolor="black",shrink=0.05))
plt.ylim(-2,2)
```

```python
mu=0.5
sd=0.3
y=mu+sd*np.random.randn(1000)
y=y[(y>0)&(y<1)]
y.sort()
x=np.arange(len(y))
```

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("Escala logaritmica")
plt.yscale("log")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x,y-y.mean())#eje logaritmico simetrico tiene sentido cuando se pinta x contra y. la media de y,
#para obtener el desplazzamiento vertical
plt.yscale("symlog",linthresh=0.01)
plt.title("Escala logaritmica simétrica")
plt.grid(True)



plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala logística")
plt.grid(True)
plt.show()
```

## CAMBIOS EN LOS EJES DE DIBUJO
Sirve para colocar los ejes en la parte central

```python
x=np.linspace(-np.pi,np.pi,256,endpoint=True)#rango de -pi a pi,256 valores del seno
S,C=np.sin(x),np.cos(x)#Se asignan a las dos variabbles las funciones, S la primera y C la segunda
plt.plot(x,S,color="green",linewidth=1.2,linestyle="-.")
plt.plot(x,C,color="blue",linewidth=1.2,linestyle="--")
plt.xlim(-4,4)
plt.ylim(-1.2,1.2)
plt.xticks(np.linspace(-4,4,9,endpoint=True))#de -a a 4, has 9 divisiones y marca los puntos finales
plt.yticks(np.linspace(-1,1,5,endpoint=True))
plt.grid(True)
plt.show()
```

```python
x=np.linspace(-np.pi,np.pi,256,endpoint=True)#rango de -pi a pi,256 valores del seno
S,C=np.sin(x),np.cos(x)#Se asignan a las dos variabbles las funciones, S la primera y C la segunda
plt.plot(x,S,color="green",linewidth=1.2,linestyle="-.")
plt.plot(x,C,color="blue",linewidth=1.2,linestyle="--")
plt.xlim(-4,4)
plt.ylim(S.min()*1.1,S.max()*1.1)
plt.xticks(np.linspace(-4,4,9,endpoint=True))#de -a a 4, has 9 divisiones y marca los puntos finales
plt.yticks(np.linspace(-1,1,5,endpoint=True))
plt.grid(True)
plt.show()
```

```python
x=np.linspace(-np.pi,np.pi,256,endpoint=True)#rango de -pi a pi,256 valores del seno
S,C=np.sin(x),np.cos(x)#Se asignan a las dos variabbles las funciones, S la primera y C la segunda
plt.plot(x,S,color="green",linewidth=1.2,linestyle="-.")
plt.plot(x,C,color="blue",linewidth=1.2,linestyle="--")
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(S.min()*1.1,S.max()*1.1)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\dfrac{\pi}{2}$',r'0',r'$\dfrac{\pi}{2}$',r'$\pi$'])#de -a a 4, has 9 divisiones y marca los puntos finales
plt.yticks(np.linspace(-1,1,3,endpoint=True))
plt.grid(True)
plt.show()
```

```python
x=np.linspace(-np.pi,np.pi,256,endpoint=True)#rango de -pi a pi,256 valores del seno
S,C=np.sin(x),np.cos(x)#Se asignan a las dos variabbles las funciones, S la primera y C la segunda
plt.plot(x,S,color="blue",linewidth=1.2,linestyle="-",label="Sin")
plt.plot(x,C,color="green",linewidth=1.2,linestyle="-",label="Cos")
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(S.min()*1.1,S.max()*1.1)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\dfrac{\pi}{2}$',' ',r'$\dfrac{\pi}{2}$',r'$\pi$'])#de -a a 4, has 9 divisiones y marca los puntos finales
plt.yticks(np.linspace(-1,1,3,endpoint=True))
plt.grid(True)

ax=plt.gca()#Da acceso a los ejes del grafico y darnos acceso
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.legend(loc='upper left')
x0=2*np.pi/3
plt.plot([x0,x0],[0,np.sin(x0)],color="blue",linewidth=2.5,linestyle="--")
plt.scatter([x0,],[np.sin(x0),],50,color="blue")
plt.annotate(r'$\sin(\dfrac{2}{3}\pi)=\dfrac{\sqrt{3}}{2}$',xy=(x0,np.sin(x0)),xycoords="data",
             xytext=(+10,+20),textcoords="offset points",
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=0.5"))#10 puntos a la derecha 30 hacia arriba

plt.show()
```
