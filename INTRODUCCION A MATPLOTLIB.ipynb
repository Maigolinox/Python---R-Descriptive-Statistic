Matplotlib tiene una sub-libreria que hace que matplot lib que es una coleccionde funciones de estilizado, la libreria es pyplot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
plt.plot(x)
plt.xlabel("Eje de abcisas")
plt.ylabel("Eje de las ordenadas")
plt.show()
```
