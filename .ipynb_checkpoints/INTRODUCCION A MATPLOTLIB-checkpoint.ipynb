Matplotlib tiene una sub-libreria que hace que matplot lib que es una coleccionde funciones de estilizado, la libreria es pyplot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
plt.plot(x)
plt.xlabel("Eje de abcisas")
plt.ylabel("Eje de las ordenadas")
plt.show()
```

Ã‰ste plot difiere del de R por que este te crea un grafico continuo, a diferencia del de R que son puntos

```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y)
```

```python
plt.plot(x,y,"ro")
plt.axis([0,8,0,18])
plt.show()
```

```python
import numpy as np
```

```python
data=np.arange(0.0,10.0,0.2)
data
```
